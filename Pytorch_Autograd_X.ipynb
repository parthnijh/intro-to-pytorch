```python
x=torch.tensor(6.7)
y=torch.tensor(0.0)
w=torch.tensor(1.0,requires_grad=True)
b=torch.tensor(0.0,requires_grad=True)
z=w*x+b
```

```python
ypred=torch.sigmoid(z)
```

```python
ypred
```

```python
criterion = nn.BCELoss()
loss = criterion(ypred, y)
```

```python
loss
```

```python
loss.backward()# computes respective derivates
w.grad
```
