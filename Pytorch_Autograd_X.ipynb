```python
x=torch.tensor(6.7)
y=torch.tensor(0.0)
w=torch.tensor(1.0,requires_grad=True)
b=torch.tensor(0.0,requires_grad=True)
z=w*x+b
```

```python
ypred=torch.sigmoid(z)
```

```python
ypred
```

```python
criterion = nn.BCELoss()
loss = criterion(ypred, y)
```

```python
loss
```

```python
loss.backward()# computes respective derivates
w.grad
```

```python
b.grad
```

```python
## clearing gradients
x=torch.tensor(2.0,requires_grad=True)


```

```python
y=x**2
y.backward()
```

```python
x.grad
x.grad.zero_()
```

```python
#disable gradient tracking
x.requires_grad_(False)

#2 detach
z=x.detach()
```

4 steps
#1 init parameters (init weight and bias)
#2 forward propagation(input->layers->output)
#3 computation of loss
#4 back propagate (update weights and biases)
